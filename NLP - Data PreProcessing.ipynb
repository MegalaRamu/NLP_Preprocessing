### word count

### character count

```python
def char_count(x):
    c = x.split()
    sp = ''.join(c)
    return sp
```

```python
df['char_count'] = df['twitts'].apply(lambda x: len(char_count(x)))
```

```python
df.head()
```

### average word length

```python
def avg_length(x):
    words = x.split()
    c = ''.join(words)
    return len(c)/len(words)
```

```python
avg_length("my name is megala")
```

```python
df['avg_length'] = df['twitts'].apply(lambda x: avg_length(x))
```

```python
df.head()
```

### stop words

```python
from spacy.lang.en.stop_words import STOP_WORDS as stopwords
```

```python
print(stopwords)
```

```python
def count_stopwords(x):
    count= 0
    sp = x.split()
    for i in sp:
        if i in stopwords:
            count+=1
    return count
```

```python
df['No. of stopwords'] = df['twitts'].apply(lambda x: count_stopwords(x))
```

```python
df.head()
```

### hashtags and mentions

```python
hashtag = []
def hastags(x):
    sp = x.split()
    for i in sp:
        if i.startswith('#'):
            hashtag.append(i)
    return len(hashtag)
```

```python
hastags("megala is frustrated because of #infosys")
```

```python
mention = []
def mentions(x):
    sp = x.split()
    for i in sp:
        if i.startswith('@'):
            mention.append(i)
    return len(mention)
```

```python
df['No. of hashtags'] = df['twitts'].apply(lambda x: hastags(x))
```

```python
df['No. of Mentions'] = df['twitts'].apply(lambda x: mentions(x))
```

```python
df.head()
```

### special chars removal

```python
def remove_punct(x):
    x = re.sub(r'[^\w ]+', "", x)
    #\w this means alphanumeric digits upper and lower cases, numbers
    #^ means anything other than alphanumeric replace it with null
    #the space at the end ensures that the waord maintain space that it has
    x = ' '.join(x.split())
    return x
```
